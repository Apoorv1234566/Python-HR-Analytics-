# Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Cleaning the Data

```python
df.isna().sum()
```

```python
df.columns
```

```python
df.head()
```

```python
df.duplicated().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.drop(['EmployeeCount'], inplace=True, axis=1)
```

```python
df = df.rename(columns={"EmployeeNumber":"EmployeeID", "NumCompaniesWorked":"PastCompaniesWorked", "Education":"Educationlevel"})
```

# Visualizing the Data

```python
num = df.select_dtypes(include =['number'])
```

```python
corr = num.corr()
plt.figure(figsize = (12,8))
s=sns.heatmap(corr,annot = True,cmap='coolwarm',fmt=".2f")
plt.title = ('correlation Map')
s.set_xticklabels(s.get_xticklabels(),rotation=45,horizontalalignment='right')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x='Attrition', data=df)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x='OverTime', data=df)
plt.show()
```

```python
plt.figure(figsize=(15,7))
plt.show('MaritalStatus')
s=sns.countplot(x='MaritalStatus',data=df)
s.set_xticklabels(s.get_xticklabels(),rotation=45,horizontalalignment='right')
```

```python
plt.figure(figsize=(15,7))
s=sns.countplot(x='JobRole',data=df)
s.set_xticklabels(s.get_xticklabels(),rotation=45,horizontalalignment='right')
```

```python
sns.set(rc={'figure.figsize':(15,7)})
plt.xticks(fontsize=12,rotation='vertical')
p=sns.barplot(x='EducationField', y='Educationlevel',data=df,palette="plasma",capsize =.2)
p.axes.set_title("Educationlevel",fontsize=20)
p.set_xticklabels(p.get_xticklabels(),rotation=45,horizontalalignment='right')
```

```python
plt.figure(figsize=(8,6))
plt.show('Department')
s=sns.countplot(x='Department',data=df)
```
